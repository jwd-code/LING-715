```python
import numpy as np 
import matplotlib
from matplotlib import pyplot as plt


file = open('C:\\Users\\Jerem\\Downloads\\perceptron_training_data.tsv')
```

```python
feats1 = []
feats2 = []
labels = []

for line in file.readlines():
    print(line)
    f1, f2, l = line.split()
    try:
        feats1.append(float(f1))
        feats2.append(float(f2))
        labels.append(int(l))
    except:
        pass
```

```python
def step_function(output):
    if output > 0:
        return 1
    else:
        return 0


def get_decision_boundary(x1, x2, weights_matrix, bias):
    w1, w2 = weights_matrix
    y1 = float((-w2*x2 - bias)/(w1))
    y2 = float((-w1*x1 - bias)/(w2))
    #return y1, y2


def decision_boundary(x1, x2, weights_matrix, bias):
    w1, w2 = weights_matrix
    
    y = (-(bias / w2) / (bias / w1))*x1 + (-bias / w2)

```

```python
epochs = 25
bias = np.random.rand(1)
weights = np.random.rand(2)
learning_rate = 0.001
```

```python
for i in range(epochs):
    correct = 0
    for x1, x2, label in zip(feats1, feats2, labels):
        x = np.array([x1, x2])
        y_hat = step_function(np.dot(x, weights))
        if y_hat != label and y_hat == 0:
            weights = weights + learning_rate*x
            bias = bias + learning_rate
        elif y_hat != label and y_hat == 1:
            weights = weights - learning_rate*x
            bias = bias - learning_rate
        elif y_hat == label:
            correct += 1
    

    print('======================EPOCH {}======================'.format(i))
    print('Correct instances:', correct)
    print('Weights', weights)
    print('Bias', bias)
    if correct == len(labels):
        print('Converged after {} epochs.'.format(i))
        break
    
    colors = ['blue', 'red']
    plt.figure(figsize=(6, 5))
    ax = plt.axes()
    ax.scatter(feats1, feats2, c=labels, cmap=matplotlib.colors.ListedColormap(colors))
    #y1, y2 = get_decision_boundary(min(feats1), max(feats1), weights, bias)

    w1 = weights[0]
    w2 = weights[1]
    c = (-bias/w2)
    slope = -(bias/w2)/(bias/w1)
    f = -bias/w1
    line_y = slope*f + c
    ax.plot(f, line_y)
```
